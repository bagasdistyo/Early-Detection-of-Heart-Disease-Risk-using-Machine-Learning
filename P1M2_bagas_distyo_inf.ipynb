Inference untuk memprediksi penyakit jantung

```python
#import library
import pandas as pd
import numpy as np
import pickle
```

```python
# Memuat model yang telah disimpan
with open('deployment/model.pkl', 'rb') as file:
    loaded_pipeline = pickle.load(file)
```

```python
# Membuat data baru untuk diuji
data = {
    "Age": [35, 60, 48, 25, 72],
    "Gender": ['Male', 'Female', 'Male', 'Female', 'Male'],
    "Blood Pressure": [130, 160, 140, 120, 150],
    "Cholesterol Level": [210, 245, 230, 190, 270],
    "Exercise Habits": ['Medium', 'Low', 'High', 'High', 'Low'],
    "Smoking": ['No', 'Yes', 'No', 'No', 'Yes'],
    "Family Heart Disease": ['No', 'Yes', 'No', 'Yes', 'No'],
    "Diabetes": ['No', 'Yes', 'No', 'No', 'Yes'],
    "BMI": [24.5, 31.2, 27.8, 22.5, 34.0],
    "High Blood Pressure": ['No', 'Yes', 'Yes', 'No', 'Yes'],
    "Low HDL Cholesterol": ['No', 'Yes', 'No', 'No', 'Yes'],
    "High LDL Cholesterol": ['No', 'Yes', 'No', 'No', 'Yes'],
    "Alcohol Consumption": ['Low', 'High', 'Medium', 'Medium', 'Low'],
    "Stress Level": ['Medium', 'High', 'Low', 'Low', 'High'],
    "Sleep Hours": [7.5, 6.0, 8.0, 8.5, 5.0],
    "Sugar Consumption": ['Medium', 'High', 'Low', 'Low', 'High'],
    "Triglyceride Level": [160, 220, 190, 150, 280],
    "Fasting Blood Sugar": [90, 120, 110, 85, 130],
    "CRP Level": [5.0, 7.5, 6.0, 4.0, 8.0],
    "Homocysteine Level": [12.0, 16.5, 13.5, 10.0, 18.0]
}

# Membuat DataFrame
new_data = pd.DataFrame(data)

# Menampilkan data baru
new_data
```

```python
# Melakukan prediksi dengan pipeline
predictions = loaded_pipeline.predict(new_data)

# Menampilkan hasil prediksi langsung (0 berarti "No" dan 1 berarti "Yes")
print("Prediksi penyakit jantung untuk data baru:", predictions)
```

Berdasarkan hasil prediksi [1 1 1 0 0], model SVM yang telah dituning memprediksi bahwa tiga individu pertama dalam data baru memiliki potensi penyakit jantung (Yes), sedangkan dua sisanya tidak (No). Hal ini sesuai dengan kekuatan utama model, yaitu kemampuan yang lebih baik dalam mendeteksi kasus penyakit jantung (kelas “Yes”) setelah proses tuning. Seperti yang ditunjukkan oleh peningkatan recall dari 0.44 menjadi 0.55 pada data uji. Artinya, model cukup baik dalam menangkap kemungkinan kasus penyakit jantung, mengurangi risiko false negative, yang penting dalam konteks medis.

Namun,  model ini memiliki precision yang rendah (0.21) untuk kelas "Yes", yang berarti bahwa banyak prediksi positif ternyata salah (false positive). Ini sejalan dengan hasil prediksi di mana 3 dari 5 sampel diklasifikasikan sebagai "Yes", yang mungkin mencakup beberapa prediksi yang tidak akurat.
